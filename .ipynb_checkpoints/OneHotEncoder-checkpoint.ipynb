```python
import random
import pandas as pd
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head(20)
```

```python
def OneHotEncoder(data):
    lst = data.whoAmI.unique().tolist()
    lst = sorted(lst)
    df = pd.DataFrame(columns=lst)
    for index, i in data.iterrows():
        for column in range(0, len(lst)):
            df.loc[index,lst[column]] = 0
            if (i['whoAmI'] == lst[column]):
                df.loc[index,lst[column]] = 1
    for column in range(0, len(lst)):
        df[lst[column]] = df[lst[column]].astype(int)
    return df
myOneHot = OneHotEncoder(data)
myOneHot
```
